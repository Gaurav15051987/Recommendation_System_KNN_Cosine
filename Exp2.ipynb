```python
df_select['artist_name'] = df_select['artist_name'].fillna('unknown_artist')
df_select['content_partner'] = df_select['content_partner'].fillna('unknown_content_partner')
df_select['ReleaseType'] = df_select['ReleaseType'].fillna('unknown_ReleaseType')
df_select['releaseyear'] = df_select['releaseyear'].fillna('unknown_releaseyear')

df_final = df_select.drop(['play_count', 'viewcount'], axis = 1)

# Removing space
df_final['album_title'] = df_final['album_title'].apply(lambda x:x.replace(" ", ""))
df_final['artist_name'] = df_final['artist_name'].apply(lambda x:x.replace(" ", ""))
df_final['genre_name'] = df_final['genre_name'].apply(lambda x:x.replace(" ", ""))
df_final['resource_type'] = df_final['resource_type'].apply(lambda x:x.replace(" ", ""))
df_final['content_partner'] = df_final['content_partner'].apply(lambda x:x.replace(" ", ""))
df_final['ReleaseType'] = df_final['ReleaseType'].apply(lambda x:x.replace(" ", ""))
df_final['releaseyear'] = df_final['releaseyear'].apply(lambda x:x.replace(" ", ""))

# Exp 16
# Creating tags column
df_final['tags'] = df_final['artist_name'] +' ' + df_final['genre_name']

# # Selecting only 3 columns
# df_final = df_final[['resource_id', 'resource_title', 'tags']]

# Convert normal case to lowercase
df_final['tags'] = df_final['tags'].apply(lambda x:x.lower())


df_final.shape
```
